```python
from embiggen.embedders import FirstOrderLINEEnsmallen
from grape.datasets.linqs import Cora, get_words_data
from embiggen import GraphVisualizer
```

```python
graph, _ = get_words_data(Cora())
```

```python
model = FirstOrderLINEEnsmallen(
    epochs=1000,
    learning_rate=0.000001
)
```

```python
embedding = model.fit_transform(graph, )
```

```python
embedding = embedding.get_node_embedding_from_index(0)
```

```python
vis = GraphVisualizer(graph, decomposition_kwargs=dict(metric="cosine"))
vis.fit_nodes(embedding)
```

```python
vis.plot_node_types()
```
