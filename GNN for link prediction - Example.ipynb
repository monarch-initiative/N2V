```python
from embiggen.edge_prediction import EdgePredictionGraphNeuralNetwork
from ensmallen.datasets.string import HomoSapiens
```

```python
string = HomoSapiens()
```

```python
train, test = string.connected_holdout(train_size=0.80, random_state=42)
```

```python
string_spine = train.get_spine(embedding_size=20)
```

```python
model = EdgePredictionGraphNeuralNetwork(
    graph=train,
    node_features=string_spine,
    node_feature_names="SPINE",
    use_node_embedding=True,
)
```

```python
model.plot()
```

```python
model.fit(
    training_graph=train,
    validation_graph=test
)
```
